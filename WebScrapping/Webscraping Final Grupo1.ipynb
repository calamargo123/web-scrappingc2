## ¿Qué es WebScraping?

El web scraping se refiere al proceso de extracción de contenidos y datos de sitios web mediante software. Por ejemplo, la mayoría de los servicios de comparación de precios utilizan web scrapers para leer la información de precios de varias tiendas online. Es un proceso dentro del Data Science que se utiliza para la extracción de datos de sitios web, normalmente en el lenguaje de programación Python y mediante machine learning, simulando cómo navegaría un ser humano en determinadas páginas web. Se Extrae los datos de un sitio web con Python implica inspeccionar la página de la URL de destino, identificar los datos que desea extraer, escribir y ejecutar el código de extracción de datos y, por último, almacenar los datos en el formato deseado. 


## ¿Para Qué Se Utiliza el Web Scraping?

El web scraping se utiliza para extraer datos de páginas web de manera automatizada. Consiste en el proceso de recopilar información de manera estructurada desde diferentes sitios web, analizando el código HTML de las páginas y extrayendo los datos relevantes. Estos datos pueden incluir texto, imágenes, tablas, enlaces y otra información disponible en la página.


El web scraping es una técnica muy útil en diversas situaciones. Algunos casos comunes de uso incluyen:

- Extracción de datos: Permite recopilar grandes cantidades de información de diferentes fuentes en un formato estructurado. Esto puede ser útil para análisis de mercado, investigación competitiva, seguimiento de precios, recopilación de datos para estudios científicos, entre otros.
- Monitoreo y seguimiento de sitios web: El scraping se puede utilizar para rastrear cambios en sitios web específicos, como actualizaciones de precios, noticias o disponibilidad de productos. Esto puede ser útil para tomar decisiones comerciales, mantenerse actualizado con la competencia o recibir alertas sobre eventos importantes.
- Agregación de contenido: Permite recopilar contenido de múltiples fuentes y combinarlo en un solo lugar. Esto se utiliza en la creación de agregadores de noticias, directorios de empresas, comparadores de precios y otros sitios web que necesitan información actualizada proveniente de diferentes fuentes.
- Análisis de datos: Los datos extraídos mediante web scraping se pueden utilizar para realizar análisis y obtener información útil. Esto incluye técnicas como el análisis de sentimientos en comentarios de clientes, el análisis de tendencias en redes sociales o la extracción de datos para entrenar modelos de aprendizaje automático.


## ¿Cómo Funciona el Scraping Web?

El funcionamiento de Web Scraping consiste en una simulación de la navegación, que una persona realizaría, ya sea implementado a bajo nivel, como en HTTP (Hypertext Transfer Protocol), o incluido en ciertos navegadores web. Ésta simulación se realiza mediante programas denominados “bots”, que crean una clonación del cliqueo, la lectura y el “copy-paste”, automatizando la tarea de búsqueda y de recolección de datos. De modo que, Web Scraping se centra en la transformación del contenido no estructurado, por lo general en formato HTML, en datos estructurados que pueden ser almacenados y analizados.

## HERRAMIENTAS PARA EL WEBSCRAPPING

Algunas de las más utilizadas son:

#### BeautifulSoup:
Es una biblioteca de Python que permite extraer datos de archivos HTML y XML. Proporciona métodos y funciones para analizar y manipular el contenido de una página web de manera sencilla.

#### Scrapy:
Es un framework de Python diseñado específicamente para el web scraping. Proporciona una estructura completa y flexible para escribir arañas (spiders) que pueden recorrer sitios web, extraer datos y seguir enlaces automáticamente.

#### Selenium:
Es una herramienta que permite la automatización de navegadores web. A través de Selenium, puedes controlar un navegador (como Chrome o Firefox) de manera programática, interactuar con elementos de la página web y extraer datos. Es útil cuando se necesita interactuar con sitios web dinámicos o que requieren autenticación.

#### Puppeteer:
Es una biblioteca de Node.js que proporciona un control de alto nivel sobre el navegador Chrome o Chromium a través de la API de DevTools Protocol. Permite realizar acciones automatizadas en el navegador, como hacer clic en elementos, completar formularios y extraer datos.
 
#### Octoparse:
Es una herramienta de web scraping visual que no requiere conocimientos de programación. Permite crear rutas de navegación en el sitio web y extraer datos mediante selecciones visuales. Es una buena opción para aquellos que no tienen experiencia en programación pero desean realizar web scraping.

```python
import bs4 as bs 
import requests
from tqdm import tqdm # Para ver cuanto demora la ejecucion del codigo
import pandas as pd
```

```python
#Request para leer la dara sin cambiar informacion
web = requests.get("http://www.iris.washington.edu/latin_am/evlist.phtml?region=mundo")
tex = web.text #pasamos a texto
#convertimos a un objeto de beautifulsoup para trabajar com html de manera eficiente
#usamos 'lxml' para manejar html mas facilmente
soup = bs.BeautifulSoup(tex, "lxml") 
#buscamos una table con el nombre de la clase
tabla = soup.find("table",{"class":"tablesorter"})
tabla
```

```python
#creamos listas vacias 
columna_0 = []
columna_3 = []
columna_5 = []

#se crea un for para navegar por cada elemento y buscar los datos
for i in tqdm(tabla.findAll("tr")[1:]):
    dato_1 = i.findAll("td")[0].text
    dato_2 = i.findAll("td")[3].text
    dato_3 = i.findAll("td")[5].text
    #agregamos los datos a cada lista
    columna_0.append(dato_1)
    columna_3.append(dato_2)
    columna_5.append(dato_3)
```

```python
#visualizamos columna 5
columna_5
```

```python
#pasamos nuestros datos a un dataframe
df = pd.DataFrame(list(zip(columna_0,columna_3,columna_5)),
                 columns=["Fecha/hora","Magnitud","Localidad"])
df.head(3)
```

```python
#vemos la dimension
df.shape
```

```python
#info
df.info()
```

#### Tenemos que arreglar magnitud y la fecha

```python
#convertimos magnitud de object a float
#convertimos Fecha/hora a un datetime
df["Magnitud"] = df["Magnitud"].astype(float)
df["Fecha/hora"] = pd.to_datetime(df["Fecha/hora"])
```

```python
df.info()
```

```python
#visuaizamos el dataframe arreglado
data_arreglo = df.set_index("Fecha/hora")
data_arreglo.head()
```
